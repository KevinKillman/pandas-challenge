```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
# print(school_data)
# print(student_data)
# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_list = school_data_complete['school_name'].unique()
school_data_complete

```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = len(school_data_complete["School ID"].unique())
total_student = len(school_data_complete['student_name'])
total_budget = school_data['budget'].sum()
avg_math = school_data_complete['math_score'].mean()
avg_read = school_data_complete['reading_score'].mean()
pass_math = round((len(school_data_complete.loc[(school_data_complete['math_score']>=70)])/total_student)*100, 2)
pass_read = round((len(school_data_complete.loc[(school_data_complete['reading_score']>=70)])/total_student)*100, 2)
pass_both = round(((len(school_data_complete.loc[(school_data_complete['reading_score']>=70) &((school_data_complete['math_score']>=70))]))/total_student)*100, 2)
district_df = pd.DataFrame({'Total Schools': total_schools,
                           'Total Students': total_student,
                           'Total Budget': total_budget,
                           'Average Math Score': avg_math,
                           'Average Reading Score':avg_read,
                           'Percent Passing Math':pass_math,
                           'Percent Passing Reading':pass_read,
                           'Percent Passing Both': [pass_both]})
district_df = district_df.style.format({'Percent Passing Math': '{0:,.2f}%',	
                                        'Percent Passing Reading': '{0:,.2f}%',	
                                        'Percent Passing Both': '{0:,.2f}%',
                                        'Total Budget': '$ {0:,.2f}',	
                                        'Average Math Score': '{0:,.2f}',	
                                        'Average Reading Score': '{0:,.2f}',
                                        'Total Students': '{:,}'})
school_data_complete
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
pass_read_df = school_data_complete.loc[(school_data_complete['reading_score']>=70)]
pass_math_df = school_data_complete.loc[(school_data_complete['math_score']>=70)]
pass_both_df = school_data_complete.loc[(school_data_complete['reading_score']>=70) &((school_data_complete['math_score']>=70))]
pass_readgroup = pass_read_df.groupby('school_name')
pass_mathgroup = pass_math_df.groupby('school_name')
pass_bothgroup = pass_both_df.groupby('school_name')



school_groupby = school_data_complete.groupby(['school_name', 'type'])
total_student_school = school_groupby['Student ID'].count()
school_data = school_data.set_index(['school_name', 'type'])
budget_school = school_data['budget']
budget_student = budget_school/total_student_school
avg_school_math = school_groupby['math_score'].mean()
avg_school_read = school_groupby['reading_score'].mean()
avg_pass_math = (pass_mathgroup['Student ID'].count()/total_student_school)*100
avg_pass_read = (pass_readgroup['Student ID'].count()/total_student_school)*100
avg_pass_both = (pass_bothgroup['Student ID'].count()/total_student_school)*100


school_df = pd.DataFrame({'Total Students':total_student_school, 
                          'Total School Budget': school_data['budget'],
                          'Per Student Budget': budget_student, 
                          'Average Math Score': avg_school_math, 
                          'Average Reading Score': avg_school_read,
                          'Percent Passing Math':avg_pass_math,
                          'Percent Passing Reading':avg_pass_read,
                          'Overall Passing':avg_pass_both})
school_data = school_data.reset_index()
school_dfstyle = school_df.style.format({'Total Students': '{0:,.0f}',	
                                    'Total School Budget': '${0:,.0f}',	
                                    'Per Student Budget': '${0:,.0f}',	
                                    'Average Math Score': '{0:,.2f}',	
                                    'Average Reading Score': '{0:,.2f}',
                                    'Percent Passing Math': '{0:,.2f}%',
                                    'Percent Passing Reading': '{0:,.2f}%',
                                    'Overall Passing':'{0:,.2f}%' })
school_dfstyle
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
bins = [0,585, 630, 645, 676]
labels1 = ['<$584',	'$585-629',	'$630-644',	'$645-675']	
school_df['Per Student Bin'] = pd.cut(school_df['Per Student Budget'],bins, labels = labels1)
group_bin = school_df.groupby('Per Student Bin')
mean_df = group_bin[['Average Math Score',
                     'Average Reading Score',
                     'Percent Passing Math',	
                     'Percent Passing Reading',	
                     'Overall Passing']].mean()
mean_df.style.format({'Average Math Score': '{0:,.2f}',	
                      'Average Reading Score': '{0:,.2f}',	
                      'Percent Passing Math': '{0:,.2f}%',	
                      'Percent Passing Reading': '{0:,.2f}%',	
                      'Overall Passing': '{0:,.2f}%'})
mean_df
```
